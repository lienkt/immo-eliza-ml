```python
import json
import csv

def json_to_csv(json_file, csv_file):
    """
    Convert a JSON file to a CSV file.

    Args:
        json_file (str): Path to the input JSON file.
        csv_file (str): Path to the output CSV file.
    """

    # Open and load JSON data from file
    with open(json_file, 'r', encoding='utf-8') as f:
        data = json.load(f)

    # Ensure the JSON structure is a list of objects
    if not isinstance(data, list):
        raise ValueError("JSON must be a list of objects")

    # Check if JSON is empty
    if len(data) == 0:
        raise ValueError("JSON file is empty")

    # Use keys of the first object as CSV header
    header = data[0].keys()

    # Open CSV file for writing
    with open(csv_file, 'w', newline='', encoding='utf-8') as f:
        writer = csv.DictWriter(f, fieldnames=header)

        # Write header row
        writer.writeheader()

        # Write each JSON object as a row in CSV
        for row in data:
            writer.writerow(row)

    print(f"Successfully converted {json_file} to {csv_file}")

filename = "../data/cleaned/clean_dataframe.json"
json_to_csv(filename, "../data/cleaned/clean_dataframe.csv")
```

```python
def clean_data(df):

    # Drop ID / leakage
    df = df.drop(columns=["property_id", "address", "price_per_m2"])

    # Fill categorical
    # cat_cols = ["property_type", "city", "province", "property_state"]
    # for col in cat_cols:
    #     df[col] = df[col].fillna("Unknown")

    # Strategy: fill categorical columns with "Unknown" (explicit, auditable) 
    # (Hint: use df.select_dtypes to select only categorical columns)
    categorical_cols = df.select_dtypes(include=["object", "category"]).columns
    df[categorical_cols] = df[categorical_cols].fillna("Unknown")


    # Fill numeric + missing flags
    num_cols = df.select_dtypes(include=["float64", "int64"]).columns

    for col in num_cols:
        df[col + "_missing"] = df[col].isna().astype(int)
        df[col] = df[col].fillna(df[col].median())

    # Strategy: fill numerical columns with the median (robust to outliers)
    # numerical_cols = df.select_dtypes(include=["number"]).columns
    # df[numerical_cols] = df[numerical_cols].fillna(df[numerical_cols].median())

    # Feature engineering
    if "build_year" in df.columns:
        df["house_age"] = 2026 - df["build_year"]

    # Check that all missing values have been handled
    print(f"Missing values remaining: {df.isnull().sum().sum()}")

    return df

clean_data(df)
```
